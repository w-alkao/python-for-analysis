```python
cdf = pd.read_excel("trees_data_cleaned.xlsx")
cdf
```

```python
cdf.isnull().sum()
```

```python
cdf.drop(columns=["Nom scientifique", "Famille"], inplace=True)
cdf
```

```python
cdf["Dbh"] = round(cdf["Circonférence (cm)"]/3.14, 2)
cdf
```

```python
def category_cime(m):
  if m < 5:
    return "[0, 4]"
  elif (m >= 5) & (m < 10):
    return "[5, 9]"
  elif (m >= 10) & (m < 15):
    return "[10, 14]"
  elif (m >= 15) & (m < 20):
    return "[15, 19]"
  elif (m >= 20):
    return "[20, +]"
  else:
    return "nothing"


def category_hauteur(m):
  if m < 5:
    return "[0, 4]"
  elif (m >= 5) & (m < 10):
    return "[5, 9]"
  elif (m >= 10) & (m < 15):
    return "[10, 14]"
  elif (m >= 15) & (m < 20):
    return "[15, 19]"
  elif (m >= 20):
    return "[20, +]"
  else:
    return "nothing"
  

def category_dbh(m):
  if m < 5:
    return "[0, 4]"
  elif (m >= 5) & (m < 25):
    return "[5, 9]"
  elif (m >= 10) & (m < 45):
    return "[10, 14]"
  elif (m >= 15) & (m < 65):
    return "[15, 19]"
  elif (m >= 65):
    return "[20, +]"
  else:
    return "nothing"
```

```python
cdf["Cime_Cat"] = cdf["Cime (m)"].map(category_cime)
cdf["Hauteur_Cat"] = cdf["Hauteur (m)"].map(category_hauteur)
cdf["Dbh_Cat"] = cdf["Dbh"].map(category_dbh)
cdf
```
